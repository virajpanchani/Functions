Functions

1. Difference between a function and a method

A function is a block of code that works independently.

A method is also a function but it belongs to an object (like list, string etc.).

Example:
len(x) → function
x.append(5) → method (belongs to list object)

2. Function arguments and parameters

Parameter → the variable written inside function definition.
Example: def add(a, b): → a, b are parameters.

Argument → actual value given when calling the function.
Example: add(5, 4) → 5, 4 are arguments.

3. Different ways to define and call a function

Normal function

Function with default value

Function with *args

Function with **kwargs

Lambda function (small one-line function)

Calling is simply: funcname()

4. Purpose of return statement

It sends back a value from the function to the place it was called.
Without return, function gives None.

5. What are iterators and how they differ from iterables

Iterable → anything you can loop over (list, tuple, string).

Iterator → object that gives one element at a time using next().

Iterable → container
Iterator → step-by-step reader

6. What are generators and how they are defined

A generator is like a function but uses yield instead of return.
It gives values one-by-one instead of storing them all.

7. Advantages of generators

Saves memory

Faster for big data

Creates values only when needed (lazy loading)

8. What is a lambda function

A small one-line function without a name.
Used for simple operations, usually inside map, filter etc.

Example: lambda x: x*x

9. Purpose of map()

It applies a function to every item in an iterable.

Example:
map(str, [1,2,3]) → turns numbers to strings.

10. Difference between map(), reduce(), filter()

map() → applies function to all items

filter() → keeps only items that match a condition

reduce() → combines all items into one value (like sum)

```python
#1. Function to return sum of even numbers in a list
def sum_even(nums):
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total
```

```python
# 2. Function that returns reverse of a string
def reverse_string(s):
    return s[::-1]
```

```python
# 3. Function that returns list of squares
def square_list(lst):
    return [x*x for x in lst]
```

```python
# 4. Function to check if number is prime (1 to 200)
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# check primes from 1 to 200
for i in range(1, 201):
    if is_prime(i):
        print(i)
```

```python
# 5. Iterator class for Fibonacci sequence
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return val
```

```python
# 6. Generator that yields powers of 2
def power_of_two(n):
    for i in range(n+1):
        yield 2 ** i
```

```python
# 7. Generator function to read file line by line
def file_reader(filename):
    with open(filename, "r") as f:
        for line in f:
            yield line.strip()
```

```python
# 8. Lambda to sort list of tuples by second element
data = [(1, 40), (3, 10), (2, 20)]
sorted_list = sorted(data, key=lambda x: x[1])
```

```python
# 9. Using map() to convert Celsius → Fahrenheit

Formula: F = C * 9/5 + 32

celsius = [0, 10, 20, 30]
fahrenheit = list(map(lambda c: (c*9/5) + 32, celsius))
```

```python
# 10. Using filter() to remove vowels from string
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, text))

print(remove_vowels("Hello Python"))
```
